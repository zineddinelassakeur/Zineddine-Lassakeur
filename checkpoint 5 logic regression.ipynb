```python
import pandas as pd 
data= pd.read_csv ("titanic-passengers.csv",sep=";")
data
data.head()
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
mean_age =round(data["Age"].mean())
mean_age 
```

```python
data["Age"].fillna(mean_age,inplace=True)
```

```python
data.isnull().sum()
```

```python
data["Cabin"].value_counts()
```

```python
data["Cabin"].fillna("G6",inplace=True)
```

```python
data.isnull().sum()
```

```python
data["Embarked"].value_counts()
```

```python
data["Embarked"].fillna("S",inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data["Sex"]=encoder.fit_transform(data["Sex"])
data
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data["Survived"]=encoder.fit_transform(data["Survived"])
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
x=data[["Age", "Sex", "SibSp", "Fare"]]
y=data["Survived"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=40)
logreg = LogisticRegression()   
logreg.fit(x_train,y_train) 
y_pred  = logreg.predict(x_test)
print("Accuracy={:.2f}".format(logreg.score(x_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
#ROC & AUC
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
fpr,tpr,thresholds=roc_curve(y_test,y_pred)
auc_logs=auc(fpr,tpr)
plt.plot(fpr,tpr,color='#3456A6',marker="^", label="ROC" % auc_logs)
plt.fill_between(fpr,tpr,color="#EA7865",label="AUC = %0.2f" % auc_logs)
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.legend(loc="lower right")
plt.show()
```
